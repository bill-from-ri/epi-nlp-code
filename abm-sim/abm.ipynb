# The model in 20 lines

`run_simulation` is a thin wrapper over the `Simulation` class. Pass a `seed` and the
run is reproducible; leave it out and one is drawn for you and printed, so any run you
like can be repeated exactly.

For the interactive version - parameter sliders, the contact network, and the epidemic
curve - run `python server.py` and open http://127.0.0.1:8000.

```python
# The whole epidemic curve, not just the terminal state.
from entities import HealthState
from simulation import Simulation

curve = [s.counts[HealthState.INFECTED.value] for s in Simulation(seed=42).run(300)]
print(f"peak infected: {max(curve)} at step {curve.index(max(curve))}")
```
